```python
from urdu import URDU_ALPHABET, URDU_DIACRITICS, URDU_EXTRA_CHARACTER, URDU_PUNCTUATIONS, URDU_DIGIT
```

```python
def removing_diacritics(text):
    for i in text:
        if i in URDU_DIACRITICS:
            text = text.replace(i, "")
    return text
    
    
```

```python
def remove_numbers(text):
    for i in text:
        if i in URDU_DIGIT:
            text = text.replace(i, "")
    return text
```

```python
def removing_extra_character(text):
    for i in text:
        if ord(i) in URDU_EXTRA_CHARACTER:
            text = text.replace(i, "")
    return text
    
```

```python
def removing_punctuations(text):
    for i in text:
        if i in URDU_PUNCTUATIONS:
            text = text.replace(i, "")
    return text
```

```python
import re
def remove_newline_tab_character(text):
    regex = re.compile(r"[\n\r\t]")
    return regex.sub("", text)
```

```python
def only_char_and_space(text):
    URDU_ALPHABET.append(' ')
    
    for i in text:
        if i not in URDU_ALPHABET:
            text = text.replace(i, "")
    return text
```
